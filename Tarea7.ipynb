```python
# Problema 3.1

sumPar=0
sumaImp=0
cuentaPar=0
num=0
i=1
while i<=270:
    num = i    #se le da el valor de i a num de lo contrario num siempre sera 0  
    if num>0:
        if  num%2!=0:
            sumPar=sumPar+cuentaPar 
            cuentaPar = cuentaPar + 1
        else: sumaImp=sumaImp+num
    i=i+1
promedioPar = sumPar / cuentaPar
print(f"El promedio de los pares es {promedioPar} y la suma de los impares es {sumaImp}")
```

```python
# Problema 3.2

band='T'
i=2
sumser=0
while i<=800:
    sumser=sumser+i
    print(i)
    if band=='T':
        band='F'
        i=i+3
    else:
        band='T'
        i=i+2
print(f"El resultado de la suma es {sumser}")
```

```python
# Problema 3.3

n=int(input("Dame un numero entero: "))
band='T'
serie = 0
i=1
while i<=n:
    if band=='T':
        serie=serie+(1/i)
        band='F'
    else: 
        serie=serie-(1/i)
        band='T'
    i=i+1
print(serie)
```

```python
# Problema 3.4

nom=0
nominas=[855,760.32,1100.2,614,2600,817.5,1280.3,687,-1]
i=0
while nominas[i] > -1:
    if nominas[i] < 1000:
        nsue=round(nominas[i]*1.15,2)
    else:
        nsue=round(nominas[i]*1.12,2)
    nom=round(nom+nsue,2)
    i=i+1
    print(f"El nuevo sueldo del trabajador es {nsue}")
print(f"La nomina es {nom}")
```
